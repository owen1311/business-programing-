## Activity 3: Debugging Complex Logic (10 minutes)

### The Mystery: Why Isn't This Working?

A student created this scholarship eligibility system, but it's not working as expected:

````python
print("=== Scholarship Eligibility Checker ===")

gpa = float(input("Enter GPA (0.0-4.0): "))
community_hours = int(input("Community service hours: "))
financial_need = input("Have financial need? (yes/no): ").lower()
essay_score = int(input("Essay score (1-10): "))

# Should be eligible if:
# - GPA 3.0+ AND (community hours 50+ OR financial need) AND essay 7+

if gpa >= 3.0 and community_hours >= 50 or financial_need == "yes" and essay_score >= 7:
    print("Eligible for scholarship!")
else:
    print("Not eligible")
````

**Your debugging process:**

**Step 1: Understand the Intent**

What is this code supposed to do? Write the logic in plain English (double-click to edit):

Should be eligible if:
GPA is 3.0 or higher

AND (community hours ≥ 50 OR financial need = yes)

AND essay score ≥ 7

**Step 2: Trace Through the Logic**

What does this condition actually check?

```python
gpa >= 3.0 and community_hours >= 50 or financial_need == "yes" and essay_score >= 7
```

Think about operator precedence. What gets evaluated first?

In Python, `and` has higher precedence than `or`. This means the `and` operations are evaluated before the `or` operations.

The condition is actually interpreted as:
` (gpa >= 3.0 and community_hours >= 50) or (financial_need == "yes" and essay_score >= 7) `

This means the scholarship is granted if:
1.  GPA is 3.0 or higher AND community hours are 50 or more
OR
2.  Financial need is 'yes' AND essay score is 7 or higher.

This deviates from the intended logic because it doesn't enforce the `gpa >= 3.0` and `essay_score >= 7` as mandatory for *all* paths to eligibility.


**Step 3: Identify the Problem (double-click to edit):**

The problem is:
The original code's logical expression is incorrectly evaluated due to Python's operator precedence. The `and` operators are evaluated before the `or` operator, leading to a condition that is `(A and B) or (C and D)` instead of `A and (B or C) and D` as intended. This means a student could be eligible solely based on financial need and essay score, even with a low GPA, or solely based on GPA and community hours, even with a low essay score.

The fix is:
To use parentheses to explicitly define the order of operations, ensuring that the 'OR' condition (`community_hours >= 50 or financial_need == "yes"`) is evaluated first, and then combined with the GPA and essay score conditions using 'AND'.

Corrected logic:
`gpa >= 3.0 and (community_hours >= 50 or financial_need == "yes") and essay_score >= 7`

````python
# Your corrected code here
print("=== Scholarship Eligibility Checker ===")

gpa = float(input("Enter GPA (0.0-4.0): "))
community_hours = int(input("Community service hours: "))
financial_need = input("Have financial need? (yes/no): ").lower()
essay_score = int(input("Essay score (1-10): "))

# Correct logic using parentheses

if gpa >= 3.0 and (community_hours >= 50 or financial_need == "yes") and essay_score >= 7:
    print("Eligible for scholarship!")
else:
    print("Not eligible")
````
