**Step 2: Design Your Algorithm** (5 minutes)


> Create a flowchart or written logic for your system (double-click to edit):

```

ELIF stress level > 7 THEN recommend take a break and relax

```

**Step 3: Implement and Test** (10 minutes)

Now write the Python code for your system. You can:

**Code your solution here:**

````python
# Your study recommendation system
study_hours = int(input("How many hours did you study today? "))
stress_level = int(input("What is your stress level (1-10)? "))
exam_days = int(input("How many days until your next exam? "))

if exam_days == 0 and study_hours < 2:
    recommendation = "Study immediately for your exam."
elif stress_level > 7:
    recommendation = "Take a break and relax before studying again."
else:
    recommendation = "Continue your normal study routine."

print("Recommendation:", recommendation)
````

## Activity 3: Debugging and Refinement (10 minutes)

### The Broken Code Challenge

Here's some code that doesn't work as intended. Your job: **Figure out what's wrong by thinking through the logic first**.

````python
student_gpa = 3.2

if student_gpa >= 2.0:
    status = "Probation"
elif student_gpa >= 3.0:
    status = "Good Standing"
elif student_gpa >= 3.5:
    status = "Dean's List"
else:
    status = "Failing"

print(f"Status: {status}")
````

**Step 1: Trace Through the Logic**

What happens when `student_gpa = 3.2`?

- Check 1: `3.2 >= 2.0`? → true
- If true, what happens next? → The program immediately assigns "Probation" and stops checking the other conditions.

**Step 2: Identify the Problem**

What's wrong with this logic? Write your analysis (double-click to edit):

The problem is: The conditions are in the wrong order.

It happens because:The program checks >= 2.0 first, so almost all GPAs above 2.0 will immediately be classified as Probation and the later conditions will never run.

**Step 3: Fix It**

How would you fix this code? Write your corrected version:

````python
# Your fixed version here
student_gpa = 3.2

if student_gpa >= 3.5:
    status = "Dean's List"
elif student_gpa >= 3.0:
    status = "Good Standing"
elif student_gpa >= 2.0:
    status = "Probation"
else:
    status = "Failing"

print(f"Status: {status}")
````
