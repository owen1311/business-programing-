## 📁 Loading Sample Data

First, let's load our sample transaction data:

````python
import pandas as pd
import os

# Define the directory and file path
data_dir = '../data'
file_path = os.path.join(data_dir, 'transactions.csv')

# Create the data directory if it doesn't exist
if not os.path.exists(data_dir):
    os.makedirs(data_dir)

# Create a dummy CSV file for demonstration if it doesn't exist
if not os.path.exists(file_path):
    dummy_data = {
        'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05'],
        'Amount': ['$50.00', '$25.50', '$120.00', '$30.75', '$75.00'],
        'Category': ['Food', 'Transport', 'Shopping', 'Food', 'Utilities'],
        'Description': ['Groceries', 'Bus fare', 'New clothes', 'Restaurant', 'Electricity bill']
    }
    dummy_df = pd.DataFrame(dummy_data)
    dummy_df.to_csv(file_path, index=False)
    print(f"Created a dummy transactions.csv at {file_path}")

# Load the sample transactions
df = pd.read_csv(file_path)

# Display the first few rows to understand our data
print("📋 Sample Transaction Data:")
print("=" * 40)
print(df.head())

print(f"\n📊 Dataset Info:")
print(f"Total transactions: {len(df)}")

# Convert 'Date' column to datetime objects for proper min/max calculation
df['Date'] = pd.to_datetime(df['Date'])
print(f"Date range: {df['Date'].min().strftime('%Y-%m-%d')} to {df['Date'].max().strftime('%Y-%m-%d')}")
print(f"Categories: {df['Category'].unique()}")
````

**Expected AI Response** (try this yourself first!):

````python
# AI-generated solution for spending analysis

# First, clean the 'Amount' column: remove dollar signs and convert to numeric
# Use errors='coerce' to turn any values that cannot be converted into NaN
df['Amount_Cleaned'] = pd.to_numeric(df['Amount'].str.replace('$', '', regex=False), errors='coerce')

# Drop rows where 'Amount_Cleaned' is NaN (invalid transactions)
df.dropna(subset=['Amount_Cleaned'], inplace=True)

# Calculate comprehensive spending summary using the cleaned column
spending_analysis = df.groupby('Category')['Amount_Cleaned'].agg(
    Total_Spent='sum',
    Avg_Amount='mean',
    Transaction_Count='count'
).round(2)

# Sort by total spending
spending_analysis = spending_analysis.sort_values('Total_Spent', ascending=False)

print("💼 Business Spending Analysis")
print("=" * 50)
print(f"{'Category':<15} {'Total':<10} {'Average':<10} {'Count':<8}")
print("-" * 50)

for category, row in spending_analysis.iterrows():
    total = row['Total_Spent']
    avg = row['Avg_Amount']
    count = row['Transaction_Count']
    print(f"{category:<15} ${total:<9.2f} ${avg:<9.2f} {count:<8.0f}")

print(f"\nGrand Total: ${spending_analysis['Total_Spent'].sum():.2f}")
````

## 📊 Step 1: Set Up Your Data

### Data Loading and Exploration

::: {.callout-tip}
## Prompt AI for this section:

*"I'm building a Smart Finance Assistant for my programming assignment. I need to load a CSV file with transaction data (columns: Date, Amount, Category, Description). The Amount column has dollar signs that need cleaning. Please write professional Python code with detailed comments that:
1. Loads the CSV data
2. Cleans and validates the data
3. Shows basic information about the dataset

````python
import pandas as pd
import os

# --- Configuration and File Path ---
# Define the directory and file path for the transaction data
data_dir = 'data'
file_name = 'sample_transactions.csv'
file_path = os.path.join(data_dir, file_name)

# Create the data directory if it doesn't exist
if not os.path.exists(data_dir):
    os.makedirs(data_dir)
    print(f"Created directory: {data_dir}/")

# Create a dummy CSV file if it doesn't exist to ensure the code can run
if not os.path.exists(file_path):
    dummy_data = {
        'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05',
                 '2023-01-06', '2023-01-07', '2023-01-08', '2023-01-09', '2023-01-10'],
        'Amount': ['$50.00', '$25.50', '$120.00', '$30.75', '$75.00',
                   '$15.20', '$88.00', '$12.50', '$90.00', '$5.00'],
        'Category': ['Food', 'Transport', 'Shopping', 'Food', 'Utilities',
                     'Food', 'Shopping', 'Transport', 'Entertainment', 'Food'],
        'Description': ['Groceries', 'Bus fare', 'New clothes', 'Restaurant', 'Electricity bill',
                        'Coffee', 'Book', 'Train ticket', 'Movie', 'Snack']
    }
    dummy_df = pd.DataFrame(dummy_data)
    dummy_df.to_csv(file_path, index=False)
    print(f"Created dummy '{file_name}' at {file_path}")

# --- 1. Load the CSV data ---
try:
    df_transactions = pd.read_csv(file_path)
    print(f"\nSuccessfully loaded '{file_name}'")
except FileNotFoundError:
    print(f"Error: The file '{file_path}' was not found. Please ensure it exists.")
    # Exit or handle gracefully if file is crucial
    df_transactions = pd.DataFrame() # Create an empty DataFrame to avoid further errors

if not df_transactions.empty:
    # --- 2. Clean and validate the data ---
    print("\n--- Initial Data Info ---")
    df_transactions.info()

    # Convert 'Date' column to datetime objects
    # errors='coerce' will turn unparseable dates into NaT (Not a Time)
    df_transactions['Date'] = pd.to_datetime(df_transactions['Date'], errors='coerce')

    # Clean 'Amount' column: remove dollar signs and convert to numeric
    # errors='coerce' will turn unparseable amounts into NaN
    df_transactions['Amount'] = df_transactions['Amount'].astype(str).str.replace('$', '', regex=False)
    df_transactions['Amount'] = pd.to_numeric(df_transactions['Amount'], errors='coerce')

    # Handle potential data issues:
    # Drop rows where 'Date' or 'Amount' could not be parsed
    initial_rows = len(df_transactions)
    df_transactions.dropna(subset=['Date', 'Amount'], inplace=True)
    if len(df_transactions) < initial_rows:
        print(f"\nWarning: {initial_rows - len(df_transactions)} rows were dropped due to invalid 'Date' or 'Amount' values.")

    # Fill any remaining NaNs in 'Category' or 'Description' with 'Unknown'
    df_transactions['Category'].fillna('Unknown', inplace=True)
    df_transactions['Description'].fillna('No Description', inplace=True)

    # --- 3. Show basic information about the dataset ---
    print("\n--- Cleaned Data Info ---")
    df_transactions.info()

    print("\n--- First 5 rows of cleaned data ---")
    display(df_transactions.head())

    print("\n--- Basic Statistics for Numeric Columns ---")
    display(df_transactions.describe())

    print(f"\nTotal transactions after cleaning: {len(df_transactions)}")
    print(f"Date range: {df_transactions['Date'].min().strftime('%Y-%m-%d')} to {df_transactions['Date'].max().strftime('%Y-%m-%d')}")
    print(f"Unique Categories: {df_transactions['Category'].unique().tolist()}")
else:
    print("DataFrame is empty, cannot proceed with cleaning and validation.")

# Store the processed DataFrame for later use in the notebook
df = df_transactions.copy() # Assign to 'df' as used in subsequent cells
````

## 📈 Step 2: Create Core Analysis Functions

**Prompt AI** to help you create the foundation of your finance assistant:

::: {.callout-tip}
## Analysis Functions Prompt:

*"For my Smart Finance Assistant, I need functions to analyze spending data. Please create well-documented functions that:
1. Calculate total spending by category
2. Find the top 5 spending categories
3. Calculate average spending per transaction
4. Identify any unusual transactions (very high or negative amounts)
Each function should have clear business explanations and error handling."*
:::

````python
import pandas as pd

# Ensure 'df' DataFrame is available from previous steps
# Assuming df is already cleaned and 'Amount' is numeric.
if 'df' not in globals() or df.empty:
    print("Error: DataFrame 'df' not found or is empty. Please run previous cells to load and clean data.")
    # Optionally, you could load a dummy df here for development if needed
    # For this exercise, we assume df is populated correctly.


def calculate_total_spending_by_category(df_transactions):
    """
    Calculates the total spending for each category.

    Args:
        df_transactions (pd.DataFrame): DataFrame with transaction data,
                                        must contain 'Category' and numeric 'Amount' columns.

    Returns:
        pd.Series: A Series with categories as index and total spending as values,
                   sorted in descending order.
    """
    if df_transactions.empty:
        print("Warning: No data to analyze for total spending by category.")
        return pd.Series()
    if 'Category' not in df_transactions.columns or 'Amount' not in df_transactions.columns:
        print("Error: DataFrame must contain 'Category' and 'Amount' columns.")
        return pd.Series()

    # Business Explanation: Understanding where the most money goes is crucial.
    # This helps identify primary expense areas.
    total_spending = df_transactions.groupby('Category')['Amount'].sum().sort_values(ascending=False)
    return total_spending


def find_top_n_spending_categories(df_transactions, n=5):
    """
    Identifies the top N categories by total spending.

    Args:
        df_transactions (pd.DataFrame): DataFrame with transaction data.
        n (int): The number of top categories to return.

    Returns:
        pd.Series: A Series of the top N categories and their total spending.
    """
    if df_transactions.empty:
        print("Warning: No data to analyze for top spending categories.")
        return pd.Series()

    total_spending = calculate_total_spending_by_category(df_transactions)
    if total_spending.empty:
        return pd.Series()

    # Business Explanation: Pinpointing the largest spending categories allows for focused
    # review and potential budget adjustments.
    top_categories = total_spending.head(n)
    return top_categories


def calculate_average_transaction_amount(df_transactions):
    """
    Calculates the average spending per transaction across all categories.

    Args:
        df_transactions (pd.DataFrame): DataFrame with transaction data.

    Returns:
        float: The average transaction amount.
    """
    if df_transactions.empty:
        print("Warning: No data to calculate average transaction amount.")
        return 0.0
    if 'Amount' not in df_transactions.columns:
        print("Error: DataFrame must contain an 'Amount' column.")
        return 0.0

    # Business Explanation: The average transaction amount can give insight into spending habits.
    # A high average might indicate fewer, larger purchases, while a low average suggests many small purchases.
    average_amount = df_transactions['Amount'].mean()
    return average_amount


def identify_unusual_transactions(df_transactions, high_threshold_multiplier=2.0, lower_bound=0):
    """
    Identifies transactions that are unusually high or negative.
    Unusually high is defined as 'Amount' greater than `high_threshold_multiplier` times the mean amount.

    Args:
        df_transactions (pd.DataFrame): DataFrame with transaction data.
        high_threshold_multiplier (float): Multiplier for the mean to set the 'unusually high' threshold.
        lower_bound (float): The lower limit for transaction amounts to be considered unusual (e.g., negative).

    Returns:
        pd.DataFrame: A DataFrame containing the unusual transactions.
    """
    if df_transactions.empty:
        print("Warning: No data to identify unusual transactions.")
        return pd.DataFrame()
    if 'Amount' not in df_transactions.columns:
        print("Error: DataFrame must contain an 'Amount' column.")
        return pd.DataFrame()

    # Business Explanation: Flagging unusual transactions helps to quickly spot potential errors,
    # fraudulent activity, or significant one-off expenses that might skew analysis.
    mean_amount = df_transactions['Amount'].mean()
    high_threshold = mean_amount * high_threshold_multiplier

    unusual_transactions = df_transactions[
        (df_transactions['Amount'] > high_threshold) |
        (df_transactions['Amount'] < lower_bound)
    ]
    return unusual_transactions


# --- Demonstration of the functions ---
print("\n--- Running Analysis Functions ---")

# 1. Total spending by category
total_by_category = calculate_total_spending_by_category(df)
print("\nTotal Spending by Category:")
print(total_by_category.to_string(float_format="${:,.2f}".format))

# 2. Top 5 spending categories
top_5_categories = find_top_n_spending_categories(df, n=5)
print("\nTop 5 Spending Categories:")
print(top_5_categories.to_string(float_format="${:,.2f}".format))

# 3. Average spending per transaction
avg_transaction = calculate_average_transaction_amount(df)
print(f"\nAverage Spending per Transaction: ${avg_transaction:,.2f}")

# 4. Unusual transactions
unusual_transactions_df = identify_unusual_transactions(df, high_threshold_multiplier=2.5, lower_bound=-0.01)
print("\nUnusual Transactions (e.g., very high or negative):")
if not unusual_transactions_df.empty:
    display(unusual_transactions_df)
else:
    print("No unusual transactions identified based on the criteria.")
````
